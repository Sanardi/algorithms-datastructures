```python
#this i came up with before learning recursion. Lets see how awesome recursion is...

def factorial(n):
    if n <= 1:
        return 1
    else:
        n_factorial = n*(n -1)
        while n > 2:
            n = n-1
            n_factorial *= (n -1)
            
            
            
        return n_factorial

#test:

start_time = time.time()

print(factorial(1000))
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
# A memoized solution
def fib_2(n, memo):
    if memo[n] is not None:
        return memo[n]
    if n == 1 or n == 2:
        result = 1
    else:
        result = fib_2(n-1, memo) + fib_2(n-2, memo)
    memo[n] = result
    return result

def fib_memo(n):
    memo = [None] * (n + 1)
    return fib_2(n, memo)

start_time = time.time()

print("40: ", fib_memo(400))

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
#memoized solution:

def fibonacci2(n, memo):
    if memo[n] is not None:
        return memo[n]
    if n == 1 or n == 2:
        result = 1
    else:
        result =  fibonacci2(n-1, memo) + fibonacci2(n-2, memo)
    memo[n] = result  
    return result

def fib_memo(n):
    memo = [None] * (n + 1)
    return fibonacci2(n, memo)
    

start_time = time.time()

print("400: ", fib_memo(400))

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
def fib_bottom_up(n):
    if n == 1 or n == 2:
        return 1
    bottom_up = [None] * (n+1)
    #print(bottom_up)
    bottom_up[1] = 1
    bottom_up[2] = 1
    for i in range(3, n+1):
        bottom_up[i] = bottom_up[i-1] + bottom_up[i-2]
    return bottom_up[n]

start_time = time.time()
print("400: ", fib_bottom_up(400))

print("--- %s seconds ---" % (time.time() - start_time))
```
